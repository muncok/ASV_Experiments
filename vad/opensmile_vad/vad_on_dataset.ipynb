### Remove segment information from file_name

shutil.move overwrites existing file  

```python
for file in files:
    from_path = os.path.join("projects/sv_experiments/vad/reddots_vad", file)
    to_name = re.sub("\.wav000.", "", file)
    to_path = os.path.join("projects/sv_experiments/vad/reddots_vad", to_name)
    try:
        shutil.move(from_path, to_path)
    except:
        pass
```

### Complement missing files 

```python
current_files = os.listdir('reddots_vad_cut_1sec/')
```

```python
missing_files = list(set(files) - set(current_files))
```

```python
missing_files = [os.path.join(file.split('_')[1],file) for file in missing_files]
```

```python
for file in missing_files:
    f = os.path.join("../dataset/reddots_r2015q4_v1/wav/", file)
    to_f = "reddots_vad_cut_1sec/"
    shutil.copy(f, to_f)
    print(file)
```

### Make directory tree

```python
current_files = os.listdir('reddots_vad_cut_1sec/')
```

```python
for file in current_files:
    subdir = file.split("_")[1]
    path = os.path.join("reddots_vad_cut_1sec/", subdir)
    try:
        if not os.path.exists(path):
            os.makedirs(path)
    except:
        pass
    shutil.move(os.path.join("reddots_vad_cut_1sec/", file), os.path.join(path,file))
```
